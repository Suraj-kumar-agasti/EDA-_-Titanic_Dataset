```python
new_age = df.Age.median()
new_age
```

```python
df.Age.fillna(new_age,inplace = True)
```

```python
df.isnull().sum()
```

```python
# Dropping cabin column due to high null values
df.drop('Cabin',axis = 1,inplace = True)
```

```python
# Dropping PassengerId as it does not contribute to survival
df.drop('Embarked',axis = 1,inplace = True)
```

```python
df.drop('PassengerId',axis = 1,inplace = True)
df.drop('Ticket',axis = 1,inplace = True)
```

```python
# creating New Fields
#1. creating new age bands to improve prediction insights
#2.Create a new feature called Family based on Parch and SibSp to get total count of family members on board
#3. Create a fare range if it helps in analysis
```

```python
df['Age_band'] = 0
df.loc[df['Age']<=1,'Age_band'] = 'Infant'
df.loc[df['Age']<=12,'Age_band'] = 'children'
df.loc[df['Age']>12,'Age_band'] = 'Adult'
df
```

```python
#fare_band
df.loc[(df['Fare'] >= 0) & (df['Fare'] <= 10), 'Fareband'] = 1
df.loc[(df['Fare'] >= 10) & (df['Fare'] <= 15), 'Fareband'] = 2
df.loc[(df['Fare'] >= 15) & (df['Fare'] <= 35), 'Fareband'] = 3
df.loc[(df['Fare'] > 35), 'Fareband'] = 3
```
